```micropython
import hub
import time
from utime import sleep_ms, ticks_ms, ticks_diff


wait = 1000
waves = 2
SPEED = 10

peopleMotor = hub.port.A.motor
nameMotor = hub.port.C.motor
peopleMotor.preset(peopleMotor.get()[0])
nameMotor.preset(nameMotor.get()[0])
peopleMotor.run_to_position(30, speed=SPEED)

beeper = hub.sound
beeper.volume(10)
freq = 2000

totalTracks = 16
totalTies = 4 * totalTracks
checkpoints = 64

tiesPerCheckpoint = int(round(totalTies / checkpoints * 2)) / 2  # rounds to nearest .5
secondsPerTie = 10 / 64


videoLength = 4
bufferLength = round(4 + secondsPerTie * tiesPerCheckpoint) # was 5
print("buffer length: " + str(bufferLength))

cycleLength = videoLength + bufferLength


def robotAction(i):
    print("starting action " + str(i + 1) + " ... ", end = "")
    for j in range(waves):
        peopleMotor.run_to_position(-30, speed=SPEED)
        
        nameMotor.run_to_position((i+1)*360*5, speed=55)

        sleep_ms(wait)
        peopleMotor.run_to_position(30, speed=SPEED)
        sleep_ms(wait)
    print("done")

    
def getSeconds(prev):
    seconds = round(ticks_diff(ticks_ms(), start_ms)/1000)
#     if seconds != prev:
#         print(seconds)
    return seconds


print("ready")
while not hub.button.right.is_pressed(): 
    sleep_ms(10)
    
sleep_ms(1500)
print("starting")
start_ms = ticks_ms()
prev = 0

for video in range(checkpoints):
    prev = getSeconds(prev)
    while (getSeconds(prev) % cycleLength) != 0:
        sleep_ms(10)
        prev = getSeconds(prev)

    beeper.beep(freq, 100, 0)
    sleep_ms(1000)
    
    robotAction(video)
    
    # pauses for bufferlength - 1 = 4 seconds
```
